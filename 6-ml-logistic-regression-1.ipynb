# ML: Logistic Regression-1

```python
import numpy as np
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
sigmoid(100000)
```

```python
sigmoid(1.99) # output will be some value between 0.5 and 1
```

```python
sigmoid(0)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
y = sigmoid(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sigmoid')
plt.grid()
plt.show()
```
